```
# import Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
########

class MulticlassLogisticRegression:
  
  # initialise alpha and iterations
  def __init__(self, alpha=0.01, iteration=100): 
    self.alpha = alpha
    self.iteration = iteration
  ####

  # sigmoid function
  def sigmoid(self, X):
    s = 1 / (1 + np.exp(-X))
    return s
  ####

  # Cost function 
  def cost_func(self, Hyp, Y):
    m = len(Y) # to get number of examples
    cost = (1 / m) * (np.sum(-Y.T.dot(np.log(Hyp)) - (1 - Y).T.dot(np.log(1 - Hyp)))) # calculate cost (1/m)* SUM(-y*log(hyp)-(1-y)*log(1-hyp))
    return cost
  ####

  # gradient descent function
  def grad_descent(self,X,Hyp,theta,Y,m):
    gradient = np.dot(X.T, (Hyp - Y)) / m  # calculate the gradient portion
    theta -= self.alpha * gradient # weight updation in thetas * gradient value calcuated in above line
    return theta
  ####

  #
  def fit_model(self, X, Y):
    print("Fitting the given dataset..")
    self.theta = []
    self.cost = []
    X = np.insert(X, 0, 1, axis=1) # adding bias (1) to the first column of X
    m = len(Y) # to get number of examples
    for i in np.unique(Y):
      #print('Descending the gradient for label type ' + str(i) + 'vs Rest')
      y_onevsall = np.where(Y == i, 1, 0) # make the first unique class (1) and rest (0) to train 3 different models iteratively
      theta = np.zeros(X.shape[1]) # initialize the theta array [0.,0.,0.,0.] for 4 variables as described in dataset
      cost = [] # initialize cost array to append 3 different costs after each model is trained
      count = 0
      for _ in range(self.iteration):
        z = X.dot(theta) # multiply X with Theta(weights)
        hyp = self.sigmoid(z) # calculate sigmoid on z
        theta = self.grad_descent(X,hyp,theta,y_onevsall,m) # calculate gradients and update weights to return theta(weights)
        cost.append(self.cost_func(hyp,y_onevsall)) # calculate and append costs
        if((count%100) == 0):
          print("count = ",count, " ",cost[-1])
        count=count+1
      self.theta.append((theta, i)) # after each model is trained we start appending new THETAS for the next model separately
      self.cost.append((cost,i)) # after each model is trained we start appending new COSTS for the next model separately
      #print("Final costs ",cost[-1])
    return self
  ####

  # prediction of all models by getting the max value from each model to see what is predicted 
  def prediction(self, X):
    X = np.insert(X, 0, 1, axis=1)
    X_predicted = [max((self.sigmoid(i.dot(theta)), c) for theta, c in self.theta)[1] for i in X ] # predicting against the testset to see what is the predictions

    return X_predicted
  ####

  # score compares the predicted label with target label to get accuracy
  def score(self,X, Y):
    score = sum(self.prediction(X) == Y) / len(Y)
    return score
  ####

  # plotting the graphs of costs respective to each class
  def plot_cost_graph(self,costh):
    for cost,c in costh   :
                plt.plot(range(len(cost)),cost,'r')
                plt.title("Graph of Cost Function of type - " + str(c) +" vs All")
                plt.xlabel("Number of Iterations")
                plt.ylabel("Cost")
                plt.show()
  ####

# 

# We are reading and processing the data provided
filename = 'iris.csv' # since this file was written in GOOGLE COLAB path was not necessary as the file was uploaded in the notebook
data = pd.read_csv(filename,sep=',')
y_data = data['species'].values  #segregating the label value from the feature values
X = data.drop(['species'],axis=1).values
scaler = StandardScaler()
X = scaler.fit_transform(X) # standardizing data

def run_model(X,y_data,learn_rate):
  scores = []
  trainscores = []
  X_train,X_test,y_train,y_test = train_test_split(X,y_data,test_size = 0.30)
  logi = MulticlassLogisticRegression(alpha=learn_rate,iteration=2000).fit_model(X_train, y_train)
  predition1 = logi.prediction(X_test)
  score1 = logi.score(X_test,y_test)
  predition2 = logi.prediction(X_train)
  score2 = logi.score(X_train,y_train)
  print("the training accuracy of the model is ",score2)
  print("the accuracy of the model is ",score1)
  scores.append(score1)
  trainscores.append(score2)
  #print(np.mean(scores)) # just for info about the average accuracy after multiple trainings
  logi.plot_cost_graph(logi.cost) # Here we ae plotting the Cost value and showing how it is depreciating close to 0 with each iteration

print("\nModel with Learnning Rate = 0.01\n\n")
run_model(X,y_data,0.01)
print("\nModel with Learnning Rate = 0.001\n\n")
run_model(X,y_data,0.001)
print("\nModel with Learnning Rate = 0.0001\n\n")
run_model(X,y_data,0.0001)
print("\nModel with Learnning Rate = 0.00005\n\n")
run_model(X,y_data,0.00005)
```
